```python
def string_toint(string):
    return int(string.split('_')[1])

for i in df.columns[:6]:
    df[i] = df[i].apply(lambda x: string_toint(x))
```

```python
df['Measures'] = df['Measures'].apply(lambda x: 0 if x=='Sell-out units' else 1)
```

```python
df.drop('ProductPackSKU_ID',axis=1,inplace=True)
```

```python
df = df[df['Value']>=0]
```

```python
df.groupby(['ProductName_ID','Point-of-Sale_ID'])['Date'].value_counts().value_counts()
```

```python
df2 = df.groupby(['ProductFamily_ID','ProductCategory_ID','ProductBrand_ID','ProductName_ID','Point-of-Sale_ID','Date','Measures'])['Value'].sum()
```

```python
df2 = df2.reset_index()
```

```python
df2.head()
```

```python
Unit = df2[df2['Measures'] == 0]['Value'].reset_index(drop=True)
Price = df2[df2['Measures'] == 1]['Value'].reset_index(drop=True)
```

```python
df2 = df2[['ProductFamily_ID','ProductCategory_ID','ProductBrand_ID','ProductName_ID','Point-of-Sale_ID','Date']].drop_duplicates(subset=['ProductName_ID','Point-of-Sale_ID','Date']).reset_index(drop=True)
```

```python
df2.head()
```

```python
df2['Unit'] = Unit
df2['Price'] = Price
```
